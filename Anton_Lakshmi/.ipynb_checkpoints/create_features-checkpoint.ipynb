```python
import os, sys
import pandas as pd
import numpy as np
from datetime import datetime
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)

path = os.getcwd() + '/data/'

log1 = pd.read_csv(path + 'student_log_{}.csv'.format(1))
initial_col_order = [col.lower() for col in list(log1.columns)]

data = pd.DataFrame()
for i in range(1, 10):
    student_log = pd.read_csv(path + 'student_log_{}.csv'.format(i))
    student_log.columns = [col.lower() for col in student_log]
    data = pd.concat([data, student_log], ignore_index=True)

data = data[initial_col_order]
data.shape
data.head()

training = pd.read_csv(path + 'training_label.csv').drop(['AveCorrect'], axis=1)
training.columns = [col.lower() for col in training.columns]
training.shape

# testing on a little dataset
df = data
df1 = df.merge(training, on='itest_id', how="left")

actions = df1[df1.isstem.notnull()]

print("training set size: ", training.shape[0], "actions for # of students: ", len(actions.itest_id.unique()))
print("sample size useless for training: ", training.shape[0] - len(actions.itest_id.unique()))

# merged training data set
stem = actions[actions.isstem == 1]
nonstem = actions[actions.isstem == 0]

stem.shape, nonstem.shape
```

### Feature 12

```python
names = list(actions.columns)
names.remove("itest_id")
names.remove("starttime")
names.remove("endtime")
names.remove("timetaken")
names.remove("isstem")
names.remove("totalfrattempted")
cut_act = actions.drop(names, axis=1)
```

```python
grouped = cut_act.groupby("itest_id")
print(grouped.ngroups)
summary = pd.DataFrame(columns=["itest_id", "num_of_q_per_day", "isstem", "n_sessions"])
for tp in grouped:
    df = tp[1]
    student_number = df["itest_id"].iloc[0]
    isstem = df["isstem"].iloc[0]
    
    # get only date from column
    dates = df['starttime'].apply(lambda x: datetime.fromtimestamp(x))
    dates = pd.to_datetime(dates)
    df['day'] = dates.dt.date
    
    # groupby day
    grouped_by_day = df.groupby("day")
#     print(grouped_by_day.ngroups)
    day_summary = pd.DataFrame(columns=["num_of_q"])
    for tl in grouped_by_day:
#         print("Dates:")
        data = tl[1]
#         print(data["day"])
        
        # number of question this day
        min_n = np.min(data["totalfrattempted"])
        max_n = np.max(data["totalfrattempted"])
        num_of_q = max_n - min_n
        
        # add this day to the table
        day_summary = day_summary.append(pd.Series({"num_of_q": num_of_q}), ignore_index=True)
        
    
    # number of sessions for a student (one day = one session)
    n_sessions = len(grouped_by_day)
    
    # add aggregated day_summary and n_days to student
    total_q = np.sum(day_summary["num_of_q"])
    days = day_summary.shape[0]
    num_of_q_per_day = total_q / float(days)
    summary = summary.append(pd.Series({"itest_id": student_number, "num_of_q_per_day": num_of_q_per_day, "isstem": isstem, "n_sessions": n_sessions}), ignore_index=True)

# add number of actions per student
# m = grouped.size()
# m = pd.DataFrame(m, columns=["n_of_actions"])
# summary = summary.set_index('itest_id').join(m)
```

### April - May feature

```python
names = list(actions.columns)
names.remove("itest_id")
names.remove("starttime")
names.remove("endtime")
names.remove("timetaken")
names.remove("isstem")
names.remove("totalfrattempted")
names.remove("sumright")
cut_act = actions.drop(names, axis=1)
```

```python
grouped = cut_act.groupby("itest_id")
print(grouped.ngroups)
tmp = pd.DataFrame(columns=["itest_id", "n_questions", "accuracy", "isstem"])
for tp in grouped:
    df = tp[1]
    student_number = df["itest_id"].iloc[0]
    isstem = df["isstem"].iloc[0]
    
    # get only day and month from column
    dates = df['starttime'].apply(lambda x: datetime.fromtimestamp(x))
    dates = pd.to_datetime(dates)
    df['month'] = dates.dt.month
    df['day'] = dates.dt.day
    
    # groupby month
    grouped_by_month = df.groupby("month")
#     print(grouped_by_day.ngroups)
#     day_summary = pd.DataFrame(columns=["num_of_q"])
    
#     print(grouped_by_month.indices.keys())

    # initialize variables
    num_of_q_apr = 0
    num_of_q_may = 0
    correct_apr = 0
    correct_may = 0

    # april
    if 4 in grouped_by_month.indices.keys():
        april = grouped_by_month.get_group(4)
        
        # number of question this month
        min_n = np.min(april["totalfrattempted"])
        max_n = np.max(april["totalfrattempted"])
        num_of_q_apr = max_n - min_n
        
        # number of correct answers
        min_c = np.min(april["sumright"])
        max_c = np.max(april["sumright"])
        correct_apr = max_c - min_c
        
        
    # may
    if 5 in grouped_by_month.indices.keys():
        may = grouped_by_month.get_group(5)
        
        # number of question this month
        min_n = np.min(may["totalfrattempted"])
        max_n = np.max(may["totalfrattempted"])
        num_of_q_may = max_n - min_n
        
        # number of correct answers
        min_c = np.min(may["sumright"])
        max_c = np.max(may["sumright"])
        correct_may = max_c - min_c
    
        
    if 4 and 5 in grouped_by_month.indices.keys():
        n_questions = num_of_q_apr + num_of_q_may
        accuracy = (correct_apr + correct_may) / float(n_questions)
    elif 4 in grouped_by_month.indices.keys():
        n_questions = num_of_q_apr
        # fix the mistake
        if n_questions == 0:
            n_questions = 1
        accuracy = correct_apr / float(n_questions)
    elif 5 in grouped_by_month.indices.keys():
        n_questions = num_of_q_may
        accuracy = correct_may / float(n_questions)
    else:
        # initialize in case of No such monthes for a student
        accuracy = -999
        n_questions = 0
    
     
#     # number of sessions for a student (one day = one session)
#     n_sessions = len(grouped_by_day)
    
    # add accuracy and n_questions to student
    tmp = tmp.append(pd.Series({"itest_id": student_number, "n_questions": n_questions, "accuracy": accuracy, "isstem": isstem}), ignore_index=True)
```
