```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
i=100
feature=pd.DataFrame(np.random.randn(i,9), columns=(['A','B','C','D','E','F','G','H','I']))
def correlation(feature):
    corr = feature.corr()  
    sns.set(style="white")
    sns.heatmap(corr, annot=True)
    plt.show()
correlation(feature)
```

```python
print(feature)
```

```python

def histogram(feature,name):
    feature1=feature[name]
    plt.hist(feature1)
    plt.show()
    
feature_name_list = feature.columns.tolist()
for feature_name in feature_name_list:
    print(feature_name)
    histogram(feature,feature_name)
```
