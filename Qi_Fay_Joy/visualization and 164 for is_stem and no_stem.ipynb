```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def all_lower(L1):
    return [s.lower() for s in L1]

train_label = pd.read_csv("../dataset/anonymized_dataset_for_ADM2017/training_label.csv")
train_label.columns = all_lower(train_label.columns.tolist())
train_label = train_label[all_lower(['ITEST_id','MCAS','isSTEM'])]
student_information = pd.read_csv("student_information.csv")
```

```python
student_information_label = pd.merge(student_information,train_label,on='itest_id',how='right')
student_information_label
```

```python
def boxplot(res,name):
    student_information_label1=student_information_label[name]
    sns.boxplot(y=student_information_label1,x='isstem',data=student_information_label)
    plt.show()
    
student_information_label_name_list = student_information_label.columns.tolist()
for student_information_label_name in student_information_label_name_list:
    print(student_information_label_name)
    boxplot(student_information_label,student_information_label_name)
```

```python
def violinplot(student_information_label,name):
    student_information_label1=student_information_label[name]
    sns.violinplot(y=student_information_label1,x='isstem',data=student_information_label)
    plt.show()
    
student_information_label_name_list = student_information_label.columns.tolist()
for student_information_label_name in student_information_label_name_list:
    print(student_information_label_name)
    violinplot(student_information_label,student_information_label_name)
```
