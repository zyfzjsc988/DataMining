```python
# polynomial transform process
def polynomial_extention(dataframe):
    poly = PolynomialFeatures(2)
    columns_name_list = list(dataframe.columns)
    
    X = dataframe.values
    new_X = poly.fit_transform(X)
    new_columns = poly.get_feature_names(columns_name_list)
    
    new_frame = pd.DataFrame(data = new_X,index = dataframe.index,columns = new_columns)
    return new_frame
```

```python
#  all data matrix
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
def all_lower(L1):
    return [s.lower() for s in L1]

training_label = pd.read_csv("../dataset/anonymized_dataset_for_ADM2017/training_label.csv")
training_label.columns = all_lower(training_label.columns.tolist())
training_label = training_label.drop(['schoolid','mcas','avecorrect'], axis=1)

skillmatrix = pd.read_csv("../dataset/question_information.csv") # question-skill-group matrix

student_information = pd.read_csv("../dataset/student_information.csv")
# import others datasets : new feature , group by skill and group by clusters
```

# feature selection process

```python
from sklearn import preprocessing
def standard_normalize(df):
    x = df.values #returns a numpy array
    scaler = preprocessing.StandardScaler()
    x_scaled = scaler.fit_transform(x)
    df = pd.DataFrame(x_scaled,columns = df.columns,index = df.index)
    return df
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid", context="talk")
rs = np.random.RandomState(7)
from minepy import MINE
from scipy.stats import pearsonr



# variance selector
def variance_selector(dataframe):
    variance = dataframe.var(axis=0)
    x = np.array(list(dataframe.columns))
    ax = sns.barplot(x, variance, palette="Set3")
    ax.set_ylabel("Variance")
    plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='center')
    plt.title('Variance comparison between nine ave- features') 
    plt.tight_layout()
    foo_fig = plt.gcf() # 'get current figure'   
    foo_fig.savefig('variance_9.eps', format='eps', dpi=1000)
    plt.show()
    
#由于MINE的设计不是函数式的，定义mic方法将其为函数式的，
# 返回一个二元组，二元组的第2项设置成固定的P值0.5
def mic(x, y):
    m = MINE()
    m.compute_score(x, y)
    return m.mic()

def information_selector(dataframe):
    v = dataframe.values
    MIC = []
    for i in range(9):
        MIC.append(mic(v[:,i],v[:,-1]))
    x = np.array(list(dataframe.columns))[:-1]
    ax = sns.barplot(x, MIC, palette="Set3")
    ax.set_ylabel("maximal information coefficient (MIC)")
    plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='center')
    plt.title('MIC comparison between nine ave- features') 
    plt.tight_layout()
    foo_fig = plt.gcf() # 'get current figure'
    foo_fig.savefig('MIC_9.eps', format='eps', dpi=1000)
    plt.show()

def pearsonr_selector(dataframe):
    v = dataframe.values
    p = []
    for i in range(9):
        p.append(pearsonr(v[:,i],v[:,-1])[0])
    x = np.array(list(dataframe.columns))[:-1]
    ax = sns.barplot(x, p, palette="Set3")
    ax.set_ylabel("pearson coefficient")
    plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='center')
    plt.title('pearson coefficient comparison between nine ave- features')
    plt.tight_layout()
    foo_fig = plt.gcf() # 'get current figure'
    foo_fig.savefig('pearson_9.eps', format='eps', dpi=1000)
    plt.show()
```
