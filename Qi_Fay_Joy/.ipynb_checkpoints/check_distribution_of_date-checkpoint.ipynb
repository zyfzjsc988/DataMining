查看做题时间和学校之间的关系

```python
import pandas as pd
import numpy as np
```

```python
stu_info = pd.read_csv("student_information_114.csv").set_index("itest_id")
labels = pd.read_csv("training_label.csv").set_index("ITEST_id")
stu_labels = labels.join(stu_info,how="inner")
stu_labels.columns.values
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# show box graft
def boxplot(res,name):
    res1 = res[name]
    sns.boxplot(y = res1,x = 'SchoolId',data=res)
    plt.show()
    

stu_labels = stu_labels[['SchoolId','average_questions_by_day', 'average_questions_by_mon',
       'counts_day_answer', 'counts_mon_answer', 'std_questions_by_day',
       'std_questions_by_mon']].dropna(axis=0, how='any')
# stu_labels.groupby('SchoolId').apply(lambda x:len(x))
stu_labels_name_list = stu_labels.columns.tolist()

for stu_labels_name in stu_labels_name_list:
    print(stu_labels_name)
    boxplot(stu_labels,stu_labels_name)
```

```python
import seaborn as sns; 
sns.set(style="ticks", color_codes=True)
g = sns.pairplot(stu_labels, hue="SchoolId", palette="husl")
```

```python
from sklearn.decomposition import PCA   

data = stu_labels[['average_questions_by_day', 'average_questions_by_mon',
       'counts_day_answer', 'counts_mon_answer', 'std_questions_by_day',
       'std_questions_by_mon']].as_matrix() 
pca=PCA(n_components=2)  
newData=pca.fit_transform(data) 
lab = np.array(stu_labels['SchoolId']).reshape(509,1)
nump = np.hstack((newData,lab))
df = pd.DataFrame(nump,columns = ['axis-1','axis-2','SchoolId'])

# g = sns.pairplot(df, hue="SchoolId", palette="husl")
sns.lmplot(x='axis-1', y='axis-2', hue='SchoolId', data=df, fit_reg=False)
```
