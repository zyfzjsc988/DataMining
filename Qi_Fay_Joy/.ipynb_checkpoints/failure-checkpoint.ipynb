```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
student_question_matrix = pd.read_csv("../dataset/person_and_question_relationship.csv")
question_information = pd.read_csv("../dataset/question_information.csv")
```

```python
skill_correctrate = pd.merge(student_question_matrix,question_information,on='problemid',how='inner')
skill_correctrate
```

```python
correctrate_larger_than_half = (skill_correctrate['correctrate']>=0.5)*1
correctrate_larger_than_half
```

```python
skill_correctrate['correctrate_larger_than_half'] = correctrate_larger_than_half
skill_correctrate
```

```python
group = skill_correctrate.groupby(['itest_id','skill','problemid'])
group_sum = group.sum()
group_sum
```

```python
group.sum()
```

```python
group_sum1 = group_sum.groupby(['itest_id','skill']).mean()
group_sum1
```

```python
group_sum2 = group_sum1.groupby(['skill','itest_id']).mean()
group_sum2
```

```python
group_sum3 = group_sum2.groupby(['skill']).sum()
group_sum3
```

```python
group_count = group_sum2.groupby(['skill']).count()
group_sum3['rate'] = group_sum3['correctrate_larger_than_half']/group_count['correctrate_larger_than_half']
group_sum3
```

```python
group_final = group_sum3.sort_values(by='rate',ascending=False)
group_final
```

```python
group_final.plot()
plt.show()
```
